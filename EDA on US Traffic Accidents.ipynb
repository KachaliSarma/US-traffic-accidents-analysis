###Start Time

- We will explore some interesting insights on Start Time.

```python
accidents_df.Start_Time
```

- In order to be able to work with Start time, we are required to convert it into datetime format using 'to_datetime' function.

```python
accidents_df.Start_Time= pd.to_datetime(accidents_df.Start_Time)
accidents_df.Start_Time
```

```python
accidents_df.Start_Time[0]
```

```python
accidents_df['Month'] = accidents_df['Start_Time'].dt.month
accidents_df['Year'] = accidents_df['Start_Time'].dt.year
accidents_df['Hour'] = accidents_df['Start_Time'].dt.hour
accidents_df['Weekday'] = accidents_df['Start_Time'].dt.weekday

#yearly data subset 
data_2016 = accidents_df[accidents_df.Start_Time.dt.year == 2016]
data_2017 = accidents_df[accidents_df.Start_Time.dt.year == 2017]
data_2018 = accidents_df[accidents_df.Start_Time.dt.year == 2018]
data_2019 = accidents_df[accidents_df.Start_Time.dt.year == 2019]
data_2020 = accidents_df[accidents_df.Start_Time.dt.year == 2020]
data_2021 = accidents_df[accidents_df.Start_Time.dt.year == 2021]
data_2017_2019 = accidents_df[(accidents_df["Year"] >= 2017) & (accidents_df["Year"] <= 2019)]
```

Hour

```python
fig, ax = plt.subplots(figsize = (15,5))
sns.countplot(x="Hour", data= accidents_df, orient = 'v', palette = "icefire_r")
plt.annotate('Morning Peak',xy=(6,350000), fontsize=12)
plt.annotate('Afternoon Peak',xy=(15,350000), fontsize=12)
plt.annotate('go to work',xy=(6,0),xytext=(3,250000),arrowprops={'arrowstyle':'-|>'}, fontsize=12)
plt.annotate('get off work',xy=(16,0),xytext=(18,250000),arrowprops={'arrowstyle':'-|>'}, fontsize=12)
plt.show()
```

- Most of accidents occure in between 6 am  to 10 am .
- Next highest percentage of accidents occure in between 3 pm to 5 pm  

Weekday

```python
fig, ax = plt.subplots(figsize = (15,5))
c = sns.countplot(x="Weekday", data=accidents_df, orient = 'v', palette = "crest_r")
plt.annotate('Weekdays',xy=(2,480000), fontsize=12)
plt.annotate('Weekend',xy=(5.1,350000), fontsize=12)
c.set_title("No. of Accidents in Weekdays")
plt.show()
```



*   Thursday is having highest accidents count.

##Summary & Conclusion

Insights:
- No data for NY.
- Less than 4% cities have more than  1000 yearly accidents.
- Over 1300 cities reported only one accident. This might be because of lack of data.
- Most of accidents occure in between 6 am to 10 am .
- The maximum accidents occurance time for weekdays is different from weekend . 
- From the analysis it seems that from tuesday to thursday ,the frequency of accidents follow approx. similar trend.
- From Nov to Dec has peak of monthly accidents count.
- There is no informations available for  Jan,2016 ,so the yearly accident counts might be vary slightly.
- Analysis showes that US/Eastern Timezone has more than 40 percentage of accidents rate.
- Location also plays important role . For example places with no bumps and traffic signals having high accidents count .
- The countplot for weather_condition depicts that the weather condition for most of the accidents were fair,followed by clear , mostly cloudly . Mostly cloudly is a resonable reason for accidents unlike fair and clear.Which implies weather condition doesnot play any important role to cause accident.
